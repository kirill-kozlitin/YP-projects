#### 2.5 Предобработка столбца *User_score*

Посмотрим на уникальные значнения в столбце:

```python
df.user_score.value_counts()
```

Помимо Nan появилась неизвестная аббревиатура **tbd**. На просторах интеренета я нашел, что подобное сокращение используется в случаях, когда описываемая величина еще не определена в силу определенных причин. Расшифровывается (чаще всего) как "To Be Determined" или "Будет определено".

Следовательно, это говорит о том, что у данных игр на момент выгрузки датафрейма, не было устоявшегося рейтинга. Мы не знаем, низкий он, или высокий. Поэтому, заменяем пропуски на Nan. Теперь, численно получим значение медианной продолжительности жизни платформы. Информации и конкретной дате релиза в наших данных нет, как и нет точной даты заката эры платформы. Но мы можем оттолкнуться от первой даты релиза игры - обычно, вместе с выходом консоли появляются на свет и первые игры. Поэтому, дату релиза можно считать достаточно точной. Закатом платформы - будем считать дату последней вышедшей игры. Таким образом, найдя разницу двух временных событий мы сможем вычислить примерную продолжительность жизни консоли.

```python
df_lifetime = df.groupby('platform', as_index=False).year_of_release.agg({'first':'min', 'last':'max'})
```

```python
df_lifetime['lifetime'] = df_lifetime['last'] - df_lifetime['first'] + 1
```

```python
df_lifetime.head()
```

Первый этап формирования необходимого датафрейма завершен. Остался неучтенным один момент. Сейчас, в нашей таблице присутствуют все виды платформ. При этом, какие-то из них все еще достаточно современны, и время их жизни еще не подходит к концу. Если мы учтем эти данные в дальнейшем вычислении, то они могут сильно исказить реальную картину. Во избежание этого, рассмотрим только те консоли, которые уже точно свое отжили. А современные отбросим. Для этого, отфильтруем данные так, чтобы в итоговый датафрейм попали только консоли, чьи игры имеют последнюю дату релиза не равную текущему 2016 году.

```python
df_lifetime_sorted = df_lifetime.query('last != 2016')
```

```python
df_lifetime_sorted
```

Построим диаграму размаха значений. Она лучше всего проиллюстрирует разброс времени жизни платформ:

```python
plt.figure(figsize=(15,2))
plt.boxplot(df_lifetime_sorted.lifetime, vert=False)
plt.title('Диаграмма размаха времени жизни платформ')
plt.xlabel('Время жизни платформы')

print('Медианное время жизни платформы:', df_lifetime_sorted.lifetime.median(), 'лет')
```

Замечание по работе над второй гипотезой натолкнуло меня на поиски проверки условий применимости t-теста. Я нашел 2 главных условия - нормальность распределения и гомоскедантичность дисперсий. Первое условие проверяем критерием Шапиро-Уилка, второе критерием Левена. Правильно ли я понимаю ход работы над проверкой гипотез?

Проверим возможность применения t-критерия для оценки:

Условия применимости t-критерия Стьюдента:
1. Нормальность распределения 
2. Однородность дисперсий


Напишем небольшую функцию, которая будет проверять эти условия применимости:

```python
def check_ttest(data_1, data_2):
    p_value = 0.05
    if st.shapiro(data_1)[1] < p_value or st.shapiro(data_2)[1] < p_value:
        print("Результат теста Шапиро-Уилка: Данные распределены НЕ нормально")
    else:
        print("Результат теста Шапиро-Уилка: Данные распределены нормально")
    if levene(data_1, data_2).pvalue < p_value:
        print("Результат теста Левена: Дисперсии НЕ однородны")
    else:
        print("Результат теста Левена: Дисперсии однородны")
```

```python
check_ttest(df_pc['user_score'], df_xone['user_score'])
```

**Условия применимости t-критерия Стьюдента не выполняются**. В таком случае воспользуемся непараметрическим критерием Манна-Уитни:

```python
mannwhitneyu(df_pc['user_score'], df_xone['user_score']).pvalue
```
